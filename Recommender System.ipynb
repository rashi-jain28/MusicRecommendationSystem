```python
# importing the libraries required for pre-processing
import pandas as panda
import numpy as np
```

## Data Preprocessing

```python
#loading the data using Pandas
colnames=['userID', 'Timestamp', 'artistid', 'artistName','trackID','trackName'] 
dataset = panda.read_csv("userid-timestamp-artid-artname-traid-traname.tsv", sep='\t', names=colnames,header=None,error_bad_lines=False)
df = dataset.drop(['Timestamp','artistid'], axis=1)
df = df.dropna(how='any',axis=0) 
df
```

```python
# removing the non-ascii characters present in the dataset
def remove_non_ascii(text):
   return ''.join(i for i in text if ord(i)<128)

df['trackName'] = df['trackName'].apply(remove_non_ascii)
df['artistName'] = df['artistName'].apply(remove_non_ascii)

df['trackName'].replace('', np.nan, inplace=True)
df['artistName'].replace('', np.nan, inplace=True)

#drop records with NA value
df = df.dropna(how='any',axis=0)

```

```python
#combining song with
import string
df['song']=df['trackName'].map(str)+" :by "+df['artistName']
```

```python
# counting the tracks played by a specific user based on the trackID 
df_temp= panda.DataFrame({'Count':df.groupby(['userID','trackID','song']).trackID.count()}).reset_index()
df_temp
```

```python
#Renaming UserIDs and TrackIDs
df_temp['user_id'] = df_temp.groupby(['userID']).ngroup()
df_temp['track_id'] = df_temp.groupby(['trackID']).ngroup()
df_temp
```

```python
lis=[]
i=0
col=df_temp.user_id.unique()
col
for oneUser in col:
    temp=df_temp[df_temp['user_id']==oneUser]
    temp['Rating'] = panda.cut(temp['Count'], bins=5,labels=False) +1
    lis.insert(i,temp)
    i+=1

final_data=panda.concat(lis)
print(final_data)
```

```python
#Dropping previous userIDs and TrackIDs
final_df= final_data.drop(['userID','trackID'], axis=1)
```
